### Завдання 1

Для трьох введених натуральних чисел (користувач вводить числа через пробіл) знайдіть мінімальне та максимальне

```python
numbers = input("Введіть три натуральних числа через пробіл: ")
list_new = numbers.split()
list_new = [int(num) for num in list_new]
max_num = list_new[0]
min_num = list_new[0]
for element in list_new:
    if element > max_num:
        max_num = element
    if element < min_num:
        min_num = element
print ('Максимальне число', max_num)
print ('Мінімальне число', min_num)
```

### Завдання 6 

Напишіть функції, що:
* рохраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```python
import random
import statistics

#генеруэмо тисячу чисел в діапазоні від 0 до 100
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]

 # Функція, що рохраховує мінімальне та максимальне значення

def min_value (list_with_random_numbers):
    return min(list_with_random_numbers)
def max_value (list_with_random_numbers):
    return max(list_with_random_numbers)

# Функція, що рохраховує середнє значення послідовності
def mean_value (list_with_random_numbers):
    return sum(list_with_random_numbers)/len(list_with_random_numbers)

# Функція, що рохраховує медіану послідовності
def median (list_with_random_numbers):
    return statistics.median(list_with_random_numbers)

# Функція, що рохраховує частоту кожної цифри в послідовності
def digit_frequency (list_with_random_numbers):
    frequency = {}
    for number in list_with_random_numbers:
        if number in frequency:
            frequency[number] += 1
        else:
            frequency[number] = 1
    return frequency

# Функція, що рохраховує середнє квадратичне відхилення
def standard_deviation (list_with_random_numbers):
    return statistics.stdev (list_with_random_numbers)

print ('Мінімальне значення', min_value (list_with_random_numbers))
print ('Максимальне значення', max_value (list_with_random_numbers))
print ('Середнє значення', mean_value (list_with_random_numbers))
print ('Медіана',median (list_with_random_numbers))
print ('Частота цифри', digit_frequency (list_with_random_numbers))
print ('Середнє квадратичне відхилення', standard_deviation (list_with_random_numbers))                   
```
