3. Вивести середнє значення та відхилення курсу за кожний місяць.

```python
import requests
from datetime import datetime

r = requests.get('https://bank.gov.ua/NBU_Exchange/exchange_site?start=20220101&end=20221231&valcode=usd&sort=exchangedate&order=desc&json').json()

# Створення словника, який буде зберігати середні значення курсу для кожного місяця
monthly_avg_rates = {}

# Проходження по кожній даті в результатах запиту
for exchangedate in r:
    if exchangedate['cc'] == 'USD':
        date_obj = datetime.strptime(exchangedate['exchangedate'], '%d.%m.%Y')
        year_month = date_obj.strftime('%Y.%m')

        if year_month in monthly_avg_rates:
            monthly_avg_rates[year_month].append(float(exchangedate['rate']))
        else:
            monthly_avg_rates[year_month] = [float(exchangedate['rate'])]

DB = {}
for year_month, rates in monthly_avg_rates.items():
    avg_rate = sum(rates) / len(rates)
    deviation = max(rates) - min(rates)
    DB[year_month] = {f"Середнє значення курсу за {year_month}: {avg_rate:.2f}, , відхилення: {deviation:.2f}"}

DB
```

### Завдання 2:

Потрібно проаналізувати всі товари на сайті: 
https://smallpacking.agrosem.ua/shop/

Створіть базу даних, що містить таблицю із такими стовбцями: 
    - артикул товару
    - найменування
    - ціна 
    - вага

```python
from bs4 import BeautifulSoup
import requests

headers = {
    'cache-control': 'max-age=0',
    'sec-ch-ua': '"Google Chrome";v="87", " Not;A Brand";v="99", "Chromium";v="87"',
    'sec-ch-ua-mobile': '?0',
    'upgrade-insecure-requests': '1',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36',
    'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
    
}

url_list = []
for page in range(1, 11):
    url = f'https://smallpacking.agrosem.ua/products/?page={page}'
    url_list.append(url)

DB = []
for url in url_list:
    r = requests.get(url, headers=headers)
    result = r.text
    soup = BeautifulSoup(result, "lxml")  
    
    for i in soup.find('div', class_='items_wrap').find_all('div', class_='item'):
        DB.append(
        {
            'name' : i.a.get('full-name'),
            'id' : i.a.get('href')[-4:],
            'price' : i.find('p', class_='price').text.strip(),
            
        }
    )


DB
```
